```python
from pyspark.sql import SparkSession

import getpass

username = getpass.getuser()

spark = SparkSession.\
builder. \
config('spark.ui.port','0'). \
config("spark.sql.warehouse.dir", f"/user/{username}/warehouse"). \
enableHiveSupport(). \
master('yarn'). \
getOrCreate()

spark
```

```python
orders_rdd = spark.sparkContext.textFile("/public/trendytech/orders/orders_1gb.csv")
```

```python
filtered = orders_rdd.filter(lambda x : x.split(",")[-1] != "PENDING_PAYMENT")
```

```python
orders_mapped = filtered.map(lambda x:(x.split(",")[2],1))
```

```python
customer_cnt = orders_mapped.reduceByKey(lambda x,y:x+y)
```

```python
result = customer_cnt.filter(lambda x:int(x[0])<=500)
```
